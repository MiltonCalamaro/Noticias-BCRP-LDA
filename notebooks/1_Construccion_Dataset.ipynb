## Load Noticias BCRP

```
df_noticias_bcrp=pd.read_pickle("../data/raw/df_noticias_bcrp_pdf.pkl")
print(df_noticias_bcrp.shape)
df_noticias_bcrp.head()
```

## Asignar etiqueta a las noticias

```
df_noticias_bcrp["Target"]=[1 for _ in range(len(df_noticias_bcrp))]
df_noticias_otras["Target"]=[0 for _ in range(len(df_noticias_otras))]
df_noticias_total=pd.concat([df_noticias_bcrp[["Texto","Target"]],df_noticias_otras[["Texto","Target"]]],axis=0)
print(df_noticias_total.shape)
df_noticias_total.head()
```

```
### ver la distribucion de los datos de forma preliminar
print(df_noticias_total["Target"].value_counts())
df_noticias_total["Target"].value_counts().plot.bar()
plt.show()
```
